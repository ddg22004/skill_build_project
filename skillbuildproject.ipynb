```python
import matplotlib.pyplot as plt
x=z["Species"]
y=z["SepalLengthCm"]
val=z["SepalWidthCm"]
spe1=z[z["Species"]=='Iris-setosa']
spe2=z[z["Species"]=='Iris-virginica']
spe3=z[z["Species"]=='Iris-versicolor']
z.plot(x='Species', y='SepalLengthCm',kind='scatter',color='b',marker="|")
#Displaying The Plot
plt.title("Species vs the Sepal Length")
plt.show()
z.plot(x='Species', y='SepalWidthCm',kind='scatter',marker="_")
#Displaying The Plot
plt.title("Sepal Width vs Species")
plt.show()
```

```python
print("Mean of Iris setosa sepal length ",sum(spe1['SepalLengthCm'])/len(spe1),"cm") #Mean value of Sepal length of Iris Sentosa
print("Mean of Iris virginica sepal length",sum(spe2['SepalLengthCm'])/len(spe2),"cm") #Mean value of Sepal length of Iris Virginica
print("Mean of Iris versicolor sepal length",sum(spe3['SepalLengthCm'])/len(spe3),"cm") #Mean value of Sepal length of Iris Versicolor
```
